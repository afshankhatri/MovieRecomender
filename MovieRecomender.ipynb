**eval()** → Built-in function that can run anything inside a string as Python code.

Powerful but dangerous if the string comes from outside (user input, file, etc.).

Example: eval("2 + 3") → 5

Example: eval("__import__('os').system('delete files')") → 💥 Dangerous.



**ast.literal_eval()** → Like eval(), but only works for safe, known data types (numbers, strings, lists, dicts, tuples, booleans, None).

Won’t run functions or system commands.

Safer to use with unknown or external input.

Example: ast.literal_eval("[1, 2, 3]") → [1, 2, 3]

Example: ast.literal_eval("__import__('os').system('delete files')") → ❌ Error

```python
# it will remove the string and bring the data inside the list 
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# given below is the general funcition ... we can use this function at all the places where just name needs to get extracted and then store in list
# .... we can use this in keyword,prodCountry etc.

def convert(obj):
    store=[]
    for i in ast.literal_eval(obj):
        store.append(i["name"])
    return store
```

**What is .apply() doing?**
.apply() in Pandas takes a function and runs it once for every value in the column.

In your case:

The column is movies["genres"]

.apply(convert) means:

For row 1, take movies["genres"][0] and give it to convert()

For row 2, take movies["genres"][1] and give it to convert()

… and so on.

```python
# this will call the function for each row of the col genres

movies["genres"] = movies["genres"].apply(convert)
movies["keywords"] = movies["keywords"].apply(convert)
movies["production_countries"]=movies["production_countries"].apply(convert) 
movies["production_companies"] =movies["production_companies"].apply(convert)
```

```python
# vectorization 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words= "english")
# max_features will keep only top 5000 most frequently occured words
# stop words will remove the stop words from the text like eg.>the a an ...and so on 
```

```python
vectors = cv.fit_transform(new_df["tag"]).toarray()
```

```python
# this will tell us which 5000 most frequent words are selected in the cv
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps  = PorterStemmer()
```

```python
def stemming_of_text(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) 
```

```python
new_df["tag"] = new_df["tag"].apply(stemming_of_text)
```

```python
# cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    # to find the index of the movie
    movie_index = new_df[new_df["title"] == movie].index[0]
    distance = similarity[movie_index]    # this is relted to the above cosine similarity... distance = 0.99 movie is more similar .... distance =0.01 movie is less similar
    movie_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x : x[1])[1:6]

    for i in movie_list :
        print(new_df.iloc[i[0]]["title"])
```

```python
# here we have to call the function by giving movie name in the param.... and our system will recommend the movies which are similar to it 
recommend("Harry Potter and the Philosopher's Stone")
```
